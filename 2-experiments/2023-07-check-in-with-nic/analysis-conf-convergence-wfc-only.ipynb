```python
# load json file
import json
with open("psp_data_a.json", "r") as f:
    psp_data = json.load(f)
```

```python
# If I use GS only, how other properties converge?

# first every property respectively, then use the max and check every property again

# phonon frequencies
prop = "phonon_frequencies"
labels = []
y_gs = {
    "wfc_cutoff": [],
    "rho_cutoff": [],
}
y_bcc = {
    "wfc_cutoff": [],
    "rho_cutoff": [],
}
y_fcc = {
    "wfc_cutoff": [],
    "rho_cutoff": [],
}
y_sc = {
    "wfc_cutoff": [],
    "rho_cutoff": [],
}
y_diamond = {
    "wfc_cutoff": [],
    "rho_cutoff": [],
}
for psp, data in psp_data.items():
    label = ".".join(psp.split(".")[0:2])
    labels.append(label)
    
    cc = 'wfc_cutoff'
    y_gs[cc].append(data["GS"][prop][cc])
    y_bcc[cc].append(data["BCC"][prop][cc])
    y_fcc[cc].append(data["FCC"][prop][cc])
    y_sc[cc].append(data["SC"][prop][cc])
    y_diamond[cc].append(data["Diamond"][prop][cc])
    
    cc = 'rho_cutoff'
    y_gs[cc].append(data["GS"][prop][cc])
    y_bcc[cc].append(data["BCC"][prop][cc])
    y_fcc[cc].append(data["FCC"][prop][cc])
    y_sc[cc].append(data["SC"][prop][cc])
    y_diamond[cc].append(data["Diamond"][prop][cc])
    
import matplotlib.pyplot as plt

plt.plot(labels, y_gs['wfc_cutoff'], label="GS-wfc", linestyle='--', marker='s', color='red')
plt.plot(labels, y_bcc['wfc_cutoff'], label="BCC-wfc", linestyle='--', marker='p', color='blue')
plt.plot(labels, y_fcc['wfc_cutoff'], label="FCC-wfc", linestyle='--', marker='>', color='green')
plt.plot(labels, y_sc['wfc_cutoff'], label="SC-wfc", linestyle='--', marker='x', color='orange')
plt.plot(labels, y_diamond['wfc_cutoff'], label="Diamond-wfc", linestyle='--', marker='o', color='purple')
plt.xticks(rotation=45)
plt.ylabel("Ry")
plt.legend()
plt.title("Convergence of phonon frequencies")
```

```python
# If I use the max of all properties, how other properties converge?

# Q1: what is the max cutoff for each property of each pseudo?

for cc in ["GS", "BCC", "FCC", "SC", "Diamond"]:
    # cc = "GS"
    print(cc)
    for psp, data in psp_data.items():
        label = ".".join(psp.split(".")[0:2])
        
        _data = data[cc]
        wfc_prop_conf_dict = {}
        for prop, prop_data in _data.items():
            wfc_prop_conf_dict[prop] = prop_data["wfc_cutoff"]
        
        wfc_max_value = max(wfc_prop_conf_dict.values())
        wfc_max_prop = [k for k, v in wfc_prop_conf_dict.items() if v == wfc_max_value]
        
        rho_prop_conf_dict = {}
        for prop, prop_data in _data.items():
            rho_prop_conf_dict[prop] = prop_data["rho_cutoff"]
        
        rho_max_value = max(rho_prop_conf_dict.values())
        rho_max_prop = [k for k, v in rho_prop_conf_dict.items() if v == rho_max_value]
        print(label, f"({wfc_max_value}, {rho_max_value}) wrt the properties ({wfc_max_prop}, {rho_max_prop})")
    
    
```

```python
# If I use the max of all properties, how other properties converge?

# Q1: what is the max cutoff for each property of each pseudo?

y_gs = {
    "wfc_cutoff": [],
    "rho_cutoff": [],
}
y_bcc = {
    "wfc_cutoff": [],
    "rho_cutoff": [],
}
y_fcc = {
    "wfc_cutoff": [],
    "rho_cutoff": [],
}
y_sc = {
    "wfc_cutoff": [],
    "rho_cutoff": [],
}
y_diamond = {
    "wfc_cutoff": [],
    "rho_cutoff": [],
}
labels = []
for psp, data in psp_data.items():
    label = ".".join(psp.split(".")[0:2])
    labels.append(label)

for cc in ["GS", "BCC", "FCC", "SC", "Diamond"]:
# for cc in ["FCC"]:
    if cc == "FCC":
        y = y_fcc
    elif cc == "BCC":
        y = y_bcc
    elif cc == "SC":
        y = y_sc
    elif cc == "Diamond":
        y = y_diamond
    elif cc == "GS":
        y = y_gs
    print(cc)
    for psp, data in psp_data.items():
        label = ".".join(psp.split(".")[0:2])
        
        _data = data[cc]
        wfc_prop_conf_dict = {}
        for prop, prop_data in _data.items():
            wfc_prop_conf_dict[prop] = prop_data["wfc_cutoff"]
        
        wfc_max_value = max(wfc_prop_conf_dict.values())
        wfc_max_prop = [k for k, v in wfc_prop_conf_dict.items() if v == wfc_max_value]
        
        rho_prop_conf_dict = {}
        for prop, prop_data in _data.items():
            rho_prop_conf_dict[prop] = prop_data["rho_cutoff"]
        
        rho_max_value = max(rho_prop_conf_dict.values())
        rho_max_prop = [k for k, v in rho_prop_conf_dict.items() if v == rho_max_value]
        print(label, f"({wfc_max_value}, {rho_max_value}) wrt the properties ({wfc_max_prop}, {rho_max_prop})")
        y["wfc_cutoff"].append(wfc_max_value)
        y["rho_cutoff"].append(rho_max_value)
    
    
```

```python
# If I use the max of all properties, how other properties converge?

# Q2: what is the max cutoff for each configuration of each pseudo?

for prop in ["phonon_frequencies", "cohesive_energy", "pressure", "bands", "delta"]:
    print(prop)
    for psp, data in psp_data.items():
        label = ".".join(psp.split(".")[0:2])
        
        wfc_prop_conf_dict = {}
        for cc, cc_data in data.items():
            try:
                wfc_prop_conf_dict[cc] = cc_data[prop]["wfc_cutoff"]
            except:
                wfc_prop_conf_dict[cc] = -1
        
        wfc_max_value = max(wfc_prop_conf_dict.values())
        wfc_max_cc = [k for k, v in wfc_prop_conf_dict.items() if v == wfc_max_value]
        # print(label, wfc_max_value, wfc_max_cc)
        
        rho_prop_conf_dict = {}
        for cc, cc_data in data.items():
            try:
                rho_prop_conf_dict[cc] = cc_data[prop]["rho_cutoff"]
            except:
                rho_prop_conf_dict[cc] = -1
        
        rho_max_value = max(rho_prop_conf_dict.values())
        rho_max_cc = [k for k, v in rho_prop_conf_dict.items() if v == rho_max_value]
        # print(label, rho_max_value, rho_max_cc)
        print(label, f"({wfc_max_value}, {rho_max_value}) wrt the confs ({wfc_max_cc}, {rho_max_cc})")
```
